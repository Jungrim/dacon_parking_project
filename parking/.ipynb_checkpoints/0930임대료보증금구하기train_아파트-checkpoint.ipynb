```python
train_df = pd.read_csv('train.csv')
age_gender_df = pd.read_csv('age_gender_info.csv')
test_df = pd.read_csv('test.csv')
```

```python
train_df = pd.DataFrame(train_df)
train_df
```

```python
train_df.isnull().sum()
```

```python
train_df = train_df.drop_duplicates()
test_df = test_df.drop_duplicates()
```

```python

# train_df['임대보증금'] = train_df['임대보증금'].astype(float)

# ValueError: could not convert string to float: '-'
# 데이터 중에 숫자, NaN 말고 "-" 기호가 들어 있음. 형변환을 위해 변경 필요.

import re # 정규표현식
arr = []
arr2 = []
for i,j in train_df[['임대보증금','임대료']].values:
    arr.append(re.sub(r"[0-9]","",str(i)))
    arr2.append(re.sub(r"[0-9]","",str(j)))
print(set(arr))
print(set(arr2))
```

```python
train_df.loc[train_df['임대보증금'] == '-','임대보증금'] = np.nan
train_df.loc[train_df['임대료'] == '-','임대료'] = np.nan

train_df['임대보증금'] = train_df['임대보증금'].astype(float)
train_df['임대료'] = train_df['임대료'].astype(float)
```

```python
train_df
```

```python
train_df.isnull().sum()
```

```python
deposit = train_df[train_df['임대보증금'].isnull()] # 304줄
```

```python
train_df[train_df['임대료'].isnull()] # 313줄
```

```python
deposit_type = deposit.groupby('공급유형')
deposit_type.count()
```

```python
dt = dict(list(deposit_type))
```

```python
# 행복주택 결측치 채우기 - 강원도 4건
a = dt['행복주택']
a['임대보증금'] = '14379000'
a['임대료'] = '72760'
a
```

```python
# 국민임대 결측치 채우기 - 부산광역시 5건, 대구광역시 3건
b = dt['국민임대']
busan = b[0:5]
busan['임대보증금']=24829040
busan['임대료']=220959
daegu = b[5:]
daegu['임대보증금']=21522460
daegu['임대료']=185410
b=pd.concat([busan,daegu])
b
```

```python
train_df=pd.concat([train_df,a,b])
# train_df = train_df.drop([2680,2681,2682,2683], axis=0)
# train_df = train_df.drop([2547,2548,2549,2550,2551], axis=0)
# train_df = train_df.drop([2906,2907,2908], axis=0)
train_df
```

```python
train_df[train_df['임대료'].isnull()] # 301줄로 줄었음!!!!
```

```python
monthly_type = monthly.groupby('공급유형')
monthly_type.count()
```

```python
mt = dict(list(monthly_type))
```

```python
# 장기전세 임대료 결측치 채우기 - 서울특별시 9건
d = mt['장기전세']
d['임대료']=0
d
```
