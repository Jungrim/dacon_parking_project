```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

import warnings
warnings.filterwarnings("ignore")

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
test_df = pd.read_csv('new_test.csv')
test_df
```

```python
monthly = test_df[test_df['임대료'].isnull()]
deposit = test_df[test_df['임대보증금'].isnull()]
```

```python
# 어느 지역에 결측치가 있는지 확인해보자
deposit_area = deposit.groupby('지역')
deposit_area.count()
```

```python
deposit['임대보증금'] = deposit['임대보증금'].astype(float)
deposit['임대료'] = deposit['임대료'].astype(float)
deposit['전용면적'] = deposit['전용면적'].astype(float)
```

```python
# 지역별로 하나씩 제곱미터당 평균 임대료 곱해서 임대료를 채우자!
```

```python
deposit_kangwon = deposit[deposit['지역'].str.contains('강원')]
deposit_kangwon['임대료'] = deposit_kangwon['전용면적'] * 10120
deposit_kangwon
```

```python
deposit_daejeon = deposit[deposit['지역'].str.contains('대전광역시')]
deposit_daejeon['임대료'] = deposit_daejeon['전용면적'] * 10640
deposit_daejeon
```

```python
deposit_busan = deposit[deposit['지역'].str.contains('부산광역시')]
deposit_busan['임대료'] = deposit_busan['전용면적'] * 21600
deposit_busan
```

```python
deposit_ulsan = deposit[deposit['지역'].str.contains('울산광역시')]
deposit_ulsan['임대료'] = deposit_ulsan['전용면적'] * 11320
deposit_ulsan
```

```python
deposit_choongnam = deposit[deposit['지역'].str.contains('충청남도')]
deposit_choongnam['임대료'] = deposit_choongnam['전용면적'] * 10000
deposit_choongnam
```
