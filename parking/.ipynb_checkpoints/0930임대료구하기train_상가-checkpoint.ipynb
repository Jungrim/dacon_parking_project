### 공공분양 결측치 0으로 채우기

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

import warnings
warnings.filterwarnings("ignore")

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
train_df = pd.read_csv('new_train.csv')
train_df
```

```python
train_df.isnull().sum()
```

```python
monthly = train_df[train_df['임대료'].isnull()]
deposit = train_df[train_df['임대보증금'].isnull()]
```

```python
deposit_type = deposit.groupby('공급유형')
deposit_type.count()
# 일단 데이터가 없는 공공분양을 전부 0으로 채워버리자!
```

```python
monthly_type = monthly.groupby('공급유형')
monthly_type.count()
# 일단 데이터가 없는 공공분양을 전부 0으로 채워버리자!
```

```python
dt = dict(list(deposit_type))
```

```python
# 공공분양 결측치 채우기 - 데이터 없으니까 전부 0으로
a = dt['공공분양']
a['임대보증금'] = '0'
a['임대료'] = '0'
a
```

```python
train_df=pd.concat([train_df,a])
```

```python
# train_df = train_df.drop([2014,2015,2016,2017,2018,2019,2020], axis=0)
train_df.isnull().sum()
```
