### 한 단지안에 공급유형이 여러 유형일 수 있는지, 자격유형이 여러 유형일 수 있는지
- 거의 한가지 유형일 듯

```python
train.groupby('단지코드').count() # 총 392개의 단지
```

```python
train.groupby('단지코드')['공급유형'].unique().value_counts() # 한 단지안에는 거의 한 공급유형이 존재
```

```python
# 한 단지안에 공긍유형 종류
a = list(train.groupby('단지코드')['공급유형'].unique().value_counts().index)
a.sort(key=len)
a
```

```python
train.groupby('단지코드')['자격유형'].unique().value_counts() # 거의 한 단지 안에는 한 자격 유형의 사람
```

```python
# 한 단지안에 자격유형 종류
b = list(train.groupby('단지코드')['자격유형'].unique().value_counts().index)
b.sort(key=len)
b
```
