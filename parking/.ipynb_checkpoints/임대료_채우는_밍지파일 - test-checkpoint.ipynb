```python
import pandas as pd
import numpy as np
```

```python
test = pd.read_csv('new_test.csv')
```

```python
dict = {'강원도':10120, '대전광역시':10640, '부산광역시':21600, '울산광역시':11320, '충청남도':10000}
```

```python
region = pd.DataFrame(dict,index=dict.keys()).T
region = region['강원도']
region['부산광역시']
```

```python
test['임대보증금'] = test['임대보증금'].astype(float)
test['임대료'] = test['임대료'].astype(float)
test['전용면적'] = test['전용면적'].astype(float)
```

```python
# train[train['임대건물구분'] == '상가']
```

```python
mapping_values = []
for r in test[test['임대건물구분'] == '상가'].values:
    print(r)
    mapping_values.append(r[6] * region[r[4]])

test.loc[test['임대건물구분']=='상가','임대료'] = mapping_values

test[test['임대건물구분'] == '상가'].head()
```

```python
test.info()
```

```python
cols = ['임대보증금','도보 10분거리 내 지하철역 수(환승노선 수 반영)','도보 10분거리 내 버스정류장 수']
test[cols] = test[cols].fillna(0)
```

```python
test.isnull().sum() # 결측치 다채웠다!
```

```python
test.to_csv('test_final.csv')
# 오류 단지코드 제거
# 임대보증금/임대료 채우기
# 버스정류장/지하철역은 결측치 전부 0으로 채움
# 중복데이터 제거
```
